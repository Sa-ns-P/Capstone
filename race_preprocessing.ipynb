```python
# cleaning

# Remove the first row with labels
geo_data = geo_dataset.iloc[1:]

# Display the DataFrame without the first row
print(geo_data)
```

```python
# create geo_data copy
geo_data_copy = geo_data.copy()

part_black_cols = ["P9_006N", "P9_013N", "P9_018N", "P9_019N", "P9_020N", "P9_021N", 
                  "P9_029N", "P9_030N", "P9_031N", "P9_032N", "P9_039N", "P9_040N", 
                  "P9_041N", "P9_042N", "P9_043N", "P9_044N", "P9_050N", "P9_051N", 
                  "P9_052N", "P9_053N", "P9_054N", "P9_055N", "P9_060N", "P9_061N", 
                  "P9_062N", "P9_063N", "P9_066N", "P9_067N", "P9_068N", "P9_071N", 
                  "P9_073N"]

# Convert selected columns to numeric type
geo_data_copy[part_black_cols] = geo_data_copy[part_black_cols].apply(pd.to_numeric, errors='coerce')

# Sum the selected columns row-wise
geo_data_copy['part_black'] = geo_data_copy[part_black_cols].sum(axis=1)

# Display the DataFrame with the new column
print(geo_data_copy)
```

```python
# change P9_005N into numeric value
geo_data_copy["P9_005N"] = pd.to_numeric(geo_data_copy["P9_005N"])

# create a new column that contains the value of P9_005N divided by part_black
geo_data_copy["white_black"] = geo_data_copy["P9_005N"] / geo_data_copy["part_black"]

print(geo_data_copy.head(5))
```

```python
# remove ZCTA5 from all values in the name column
geo_data_copy['NAME'] = geo_data_copy['NAME'].str.replace('ZCTA5 ', '')

print(geo_data_copy.head(5))
```

```python
# DIVERSITY INDEX

# change columns needed into numeric value
geo_data_copy["P9_002N"] = pd.to_numeric(geo_data_copy["P9_002N"])
geo_data_copy["P9_001N"] = pd.to_numeric(geo_data_copy["P9_001N"])
geo_data_copy["P9_005N"] = pd.to_numeric(geo_data_copy["P9_005N"])
geo_data_copy["P9_006N"] = pd.to_numeric(geo_data_copy["P9_006N"])
geo_data_copy["P9_007N"] = pd.to_numeric(geo_data_copy["P9_007N"])
geo_data_copy["P9_008N"] = pd.to_numeric(geo_data_copy["P9_008N"])
geo_data_copy["P9_009N"] = pd.to_numeric(geo_data_copy["P9_009N"])
geo_data_copy["P9_010N"] = pd.to_numeric(geo_data_copy["P9_010N"])
geo_data_copy["P9_011N"] = pd.to_numeric(geo_data_copy["P9_011N"])

# Create a new column for the proportion of Hispanic/Latino
geo_data_copy['prop_hisp_latino'] = geo_data_copy['P9_002N'] / geo_data_copy['P9_001N']

# Create a new column for the proportion of White alone
geo_data_copy['prop_white_alone'] = geo_data_copy['P9_005N'] / geo_data_copy['P9_001N']

# Create a new column for the proportion of Black alone
geo_data_copy['prop_black_alone'] = geo_data_copy['P9_006N'] / geo_data_copy['P9_001N']

# Create a new column for the proportion of American Indian/Alaska Native alone
geo_data_copy['prop_amer_ind_alone'] = geo_data_copy['P9_007N'] / geo_data_copy['P9_001N']

# Create a new column for the proportion of Asian alone
geo_data_copy['prop_asian_alone'] = geo_data_copy['P9_008N'] / geo_data_copy['P9_001N']

# Create a new column for the proportion of Native Hawaiian/Other Pacific Islander alone
geo_data_copy['prop_native_hawaiian_alone'] = geo_data_copy['P9_009N'] / geo_data_copy['P9_001N']

# Create a new column for the proportion of Some Other Race (SOR) alone
geo_data_copy['prop_sor_alone'] = geo_data_copy['P9_010N'] / geo_data_copy['P9_001N']

# Create a new column for the proportion of Two or More Races alone
geo_data_copy['prop_two_plus_alone'] = geo_data_copy['P9_011N'] / geo_data_copy['P9_001N']

# print(geo_data_copy.sample(5))
```

```python
# Calculate the squares of proportions for each racial group
geo_data_copy['hisp_latino_sq'] = geo_data_copy['prop_hisp_latino'] ** 2
geo_data_copy['white_sq'] = geo_data_copy['prop_white_alone'] ** 2
geo_data_copy['black_sq'] = geo_data_copy['prop_black_alone'] ** 2
geo_data_copy['amer_ind_sq'] = geo_data_copy['prop_amer_ind_alone'] ** 2
geo_data_copy['asian_sq'] = geo_data_copy['prop_asian_alone'] ** 2
geo_data_copy['nhpi_sq'] = geo_data_copy['prop_native_hawaiian_alone'] ** 2
geo_data_copy['sor_sq'] = geo_data_copy['prop_sor_alone'] ** 2
geo_data_copy['two_plus_sq'] = geo_data_copy['prop_two_plus_alone'] ** 2

# Sum up the squared proportions for all racial groups
geo_data_copy['sum_sq'] = geo_data_copy[['hisp_latino_sq', 'white_sq', 'black_sq', 
                                         'amer_ind_sq', 'asian_sq', 'nhpi_sq', 
                                         'sor_sq', 'two_plus_sq']].sum(axis=1)

# Calculate diversity index
geo_data_copy['diversity_index'] = 1 - geo_data_copy['sum_sq']

# Drop intermediate columns
geo_data_copy.drop(columns=['hisp_latino_sq', 'white_sq', 'black_sq', 
                            'amer_ind_sq', 'asian_sq', 'nhpi_sq', 
                            'sor_sq', 'two_plus_sq', 'sum_sq'], inplace=True)

# Print the DataFrame to see the diversity index column
print(geo_data_copy[['NAME', 'diversity_index']].sample(5))
```

```python
# keep only columns NAME, white_black, diversity_index
final = geo_data_copy[['NAME', 'white_black', 'diversity_index']]

#final.to_csv('final.csv', index=False)
```

```python
# upload Faces and Names dataset

faces_dataset = pd.read_csv('/Users/kristivanmeter/Desktop/MIDS/Capstone/unique_participants_newfaces.csv')
faces_dataset
```

```python
# make copy of faces_dataset
faces_dataset = faces_dataset.copy()

# Add '0' as the first digit to ensure all values have 5 digits in Zip
faces_dataset.loc[faces_dataset["Zip"].astype(str).str.len() < 5, "Zip"] = faces_dataset["Zip"].astype(str).str.zfill(5)


# reset index after filtering
faces_dataset.reset_index(drop=True, inplace=True)
```

```python
# Convert 'NAME' column in final and 'Zip' column in faces_dataset to int64
final['NAME'] = final['NAME'].astype('int64').copy()
faces_dataset['Zip'] = faces_dataset['Zip'].astype('int64').copy()
```

```python

# Merge datasets based on 'Zip' column with a left join
merged_df = pd.merge(faces_dataset, final, left_on='Zip', right_on='NAME', how='left')

# Update 'white_black' values where there's a match
merged_df['white_black'] = merged_df['white_black'].fillna(0)  # Fill NaN values with 0

# Drop the extra 'NAME' column
merged_df.drop(columns=['NAME'], inplace=True)

merged_df
```

```python
# write merged data
merged_df.to_csv('merged_faces.csv', index=False)
```
